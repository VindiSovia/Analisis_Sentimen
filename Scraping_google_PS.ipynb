```python
from google_play_scraper import app, Sort, reviews
from datetime import datetime  # Import datetime for date handling

import pandas as pd

import numpy as np
```

```python
#Scrape desired number of reviews
#Run kode ini jika ingin scrape data dengan jumlah tertentu. Ganti (misal, ingin scrape sejumlah 1000, maka ganti kode , count = 1000 )

from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.shopee.id',
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT you can use Sort.NEWEST to get newst reviews
    count=5000, # defaults to 100
    filter_score_with=None # defaults to None(means all score) Use 1 or 2 or 3 or 4 or 5 to select certain score
)
```

```python
df_busu = pd.DataFrame(np.array(result),columns=['reviews'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('reviews').tolist()))

df_busu.head()
```

```python
# Filter baris berdasarkan tahun 2023
df_2023 = df_busu[df_busu['at'].dt.year == 2023]

# Pilih kolom yang ingin Anda tampilkan
new_df = df_2023[['userName', 'score', 'at', 'content']]

# Urutkan dataframe berdasarkan tanggal (dari yang terbaru)
sorted_df = new_df.sort_values(by='at', ascending=False)

# Inisialisasi variabel untuk menyimpan data teratas
top_125_df = pd.DataFrame()

# Inisialisasi variabel untuk menghitung total data yang telah diambil
total_rows_taken = 0

# Loop melalui setiap bulan dalam tahun 2023
for month in range(1, 13):
    # Filter data untuk bulan tertentu
    df_month = sorted_df[(sorted_df['at'].dt.month == month)]

    # Hitung jumlah data yang akan diambil untuk bulan ini
    if total_rows_taken + 15 <= 125:
        num_rows_to_take = min(15, len(df_month))
    else:
        num_rows_to_take = min(125 - total_rows_taken, len(df_month))

    # Ambil data sesuai dengan batasan jumlah data per bulan
    top_125_df = pd.concat([top_125_df, df_month.head(num_rows_to_take)], ignore_index=True)

    # Perbarui jumlah total data yang telah diambil
    total_rows_taken += num_rows_to_take

    # Jika jumlah total data yang telah diambil mencapai 125, keluar dari loop
    if total_rows_taken >= 125:
        break

# Pastikan jumlah data teratas tetap 125
top_125_df = top_125_df.head(125)
```
