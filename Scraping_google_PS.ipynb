```python
from google_play_scraper import app, Sort, reviews
from datetime import datetime  # Import datetime for date handling

import pandas as pd

import numpy as np
```

```python
# Scrape All available reviews
#(DONT RUN THIS CELL!!! RUN ONLY IF YOU WANT TO SCRAPE ALL AVAILABLE REVIEWS)
#JANGAN DI RUN KODE INI, KECUALI ANDA INGIN SCRAPE SEMUA REVIEWS SEKALIGUS

from google_play_scraper import Sort, reviews_all

result = reviews_all(
    'com.bukalapak.android',
    sleep_milliseconds=0, # defaults to 0
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT , you can use Sort.NEWEST to get newst reviews
)
```

```python
#Scrape desired number of reviews
#Run kode ini jika ingin scrape data dengan jumlah tertentu. Ganti (misal, ingin scrape sejumlah 1000, maka ganti kode , count = 1000 )

from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.tokopedia.tkpd',
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT you can use Sort.NEWEST to get newst reviews
    count=5000, # defaults to 100
    filter_score_with=None # defaults to None(means all score) Use 1 or 2 or 3 or 4 or 5 to select certain score
)
```

```python
df_busu = pd.DataFrame(np.array(result),columns=['reviews'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('reviews').tolist()))

df_busu.head()
```

```python
len(df_busu.index) #count the number of data we got
```

```python
df_busu[['userName', 'score','at', 'content']].head()  #preview userName, rating, date-time, and reviews only
```

```python
#Run This Code to Sort the Data By Date

new_df = df_busu[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=True) #Sort by Newst, change to True if you want to sort by Oldest.
sorted_df.head()
```

```python
# Filter baris berdasarkan tahun 2023
df_2023 = df_busu[df_busu['at'].dt.year == 2023]

# Pilih kolom yang ingin Anda tampilkan
new_df = df_2023[['userName', 'score', 'at', 'content']]

# Urutkan dataframe berdasarkan tanggal (dari yang terbaru)
sorted_df = new_df.sort_values(by='at', ascending=False)

# Buat kolom baru 'month' yang berisi bulan dari kolom 'at'
sorted_df['month'] = sorted_df['at'].dt.month

# Buat grup berdasarkan bulan
grouped_df = sorted_df.groupby('month')

# Ambil 17 data teratas dari setiap grup
result_df = grouped_df.head(17)

# Tampilkan dataframe yang berisi 17 data teratas setiap bulannya
print(result_df)
```

```python
my_df = sorted_df[['userName', 'score','at', 'content']] #get userName, rating, date-time, and reviews only
```
