```python
from app_store_scraper import AppStore

import pandas as pd

import numpy as np
```

```python
from pprint import pprint

reviews_app = AppStore(country="id", app_name="lazada-indonesia", app_id="785385147")
reviews_app.review(how_many=100)

pprint(reviews_app.reviews_count)
```
