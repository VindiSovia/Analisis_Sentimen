```python
from app_store_scraper import AppStore

import pandas as pd

import numpy as np

import json, os, uuid
```

```python
from app_store_scraper import AppStore
from pprint import pprint

reviews_app = AppStore(country="id", app_name="tokopedia", app_id="1001394201")
reviews_app.review(how_many=5000)

pprint(reviews_app.reviews)
pprint(reviews_app.reviews_count)
```

```python
# Convert reviews to DataFrame
df_busu = pd.DataFrame(reviews_app.reviews)

# Print columns of df_busu
print(df_busu.columns)
```

```python
df_busu.head()
```

```python
df_busu[['date', 'review','rating', 'title', 'userName']].head()
```

```python
# Filter baris berdasarkan tahun 2023
df_2023 = df_busu[df_busu['date'].dt.year == 2023]

# Pilih kolom yang ingin Anda tampilkan
new_df = df_2023[['date', 'review','rating', 'title', 'userName']]

# Urutkan dataframe berdasarkan tanggal (dari yang terbaru)
sorted_df = new_df.sort_values(by='date', ascending=False)

# Buat kolom baru 'month' yang berisi bulan dari kolom 'at'
sorted_df['month'] = sorted_df['date'].dt.month

# Buat grup berdasarkan bulan
grouped_df = sorted_df.groupby('month')

# Ambil 17 data teratas dari setiap grup
result_df = grouped_df.head(17)

# Tampilkan dataframe yang berisi 17 data teratas setiap bulannya
print(result_df)
```

```python
my_df = sorted_df[['userName', 'rating','date', 'review']]

my_df.head()
```

```python
my_df.to_csv("TokopediaPS1.csv", index = False)
```
