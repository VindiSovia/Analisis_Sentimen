```python
from app_store_scraper import AppStore

import pandas as pd

import numpy as np
```

```python
import time
import requests

def make_request():
    url = "https://apps.apple.com/id/app/tokopedia/id1001394201?l=id"
    try:
        response = requests.get(url)
        # Check for 429 status code
        if response.status_code == 429:
            # Implement exponential backoff
            retry_after = int(response.headers.get('Retry-After', 5))
            print(f"Received 429. Retrying after {retry_after} seconds...")
            time.sleep(retry_after)
            make_request()  # Retry the request
        else:
            # Handle other response codes or return response
            return response
    except Exception as e:
        print(f"Error occurred: {e}")

# Call the function to make the request
make_request()
```

```python
# Filter baris berdasarkan tahun 2023
df_2023 = df_busu[df_busu['date'].dt.year == 2023]

# Pilih kolom yang ingin Anda tampilkan
new_df = df_2023[['userName', 'rating','date', 'review']]

# Urutkan dataframe berdasarkan tanggal (dari yang terbaru)
sorted_df = new_df.sort_values(by='date', ascending=False)

# Inisialisasi variabel untuk menyimpan data teratas
top_125_df = pd.DataFrame()

# Inisialisasi variabel untuk menghitung total data yang telah diambil
total_rows_taken = 0

# Loop melalui setiap bulan dalam tahun 2023
for month in range(1, 13):
    # Filter data untuk bulan tertentu
    df_month = sorted_df[(sorted_df['date'].dt.month == month)]

    # Hitung jumlah data yang akan diambil untuk bulan ini
    if total_rows_taken + 15 <= 125:
        num_rows_to_take = min(15, len(df_month))
    else:
        num_rows_to_take = min(125 - total_rows_taken, len(df_month))

    # Ambil data sesuai dengan batasan jumlah data per bulan
    top_125_df = pd.concat([top_125_df, df_month.head(num_rows_to_take)], ignore_index=True)

    # Perbarui jumlah total data yang telah diambil
    total_rows_taken += num_rows_to_take

    # Jika jumlah total data yang telah diambil mencapai 125, keluar dari loop
    if total_rows_taken >= 125:
        break

# Pastikan jumlah data teratas tetap 125
top_125_df = top_125_df.head(125)
```

```python
# Filter data untuk tahun 2023 dan 2024
df_2023_2024 = df_busu[(df_busu['date'].dt.year == 2023) | (df_busu['date'].dt.year == 2024)]

# Pilih kolom yang ingin Anda tampilkan
new_df = df_2023_2024[['userName', 'rating','date', 'review']]

# Urutkan dataframe berdasarkan tanggal (dari yang terbaru)
sorted_df = new_df.sort_values(by='date', ascending=False)

# Inisialisasi variabel untuk menyimpan data teratas
top_125_df = pd.DataFrame()

# Inisialisasi variabel untuk menghitung total data yang telah diambil
total_rows_taken = 0

# Loop melalui setiap bulan dalam tahun 2023
for month in range(1, 13):
    # Filter data untuk bulan tertentu
    df_month = sorted_df[(sorted_df['date'].dt.month == month)]

    # Hitung jumlah data yang akan diambil untuk bulan ini
    if total_rows_taken + 15 <= 125:
        num_rows_to_take = min(15, len(df_month))
    else:
        num_rows_to_take = min(125 - total_rows_taken, len(df_month))

    # Ambil data sesuai dengan batasan jumlah data per bulan
    top_125_df = pd.concat([top_125_df, df_month.head(num_rows_to_take)], ignore_index=True)

    # Perbarui jumlah total data yang telah diambil
    total_rows_taken += num_rows_to_take

    # Jika jumlah total data yang telah diambil mencapai 125, keluar dari loop
    if total_rows_taken >= 125:
        break

# Pastikan jumlah data teratas tetap 125
top_125_df = top_125_df.head(125)

# Jika jumlah data kurang dari 125, tambahkan dengan data Januari 2024
if len(top_125_df) < 125:
    # Filter data untuk bulan Januari 2024
    df_jan_2024 = sorted_df[(sorted_df['date'].dt.year == 2024) & (sorted_df['date'].dt.month == 1)]

    # Hitung jumlah data yang perlu ditambahkan
    num_rows_to_add = min(125 - len(top_125_df), len(df_jan_2024))

    # Ambil data dari bulan Januari 2024
    top_125_df = pd.concat([top_125_df, df_jan_2024.head(num_rows_to_add)], ignore_index=True)
```
